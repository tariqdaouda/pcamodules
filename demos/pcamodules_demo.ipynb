```python
import scanpy as sc
import pcamodules.pcamodules as pcam
```

# Setup

```python
adata = sc.datasets.pbmc3k_processed()
```

```python
sc.pp.pca(adata)
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.tl.leiden(adata)
```

```python
sc.pl.umap(adata, color=["leiden"])
```

```python
finder = pcam.PCAModules(max_iter=1000)
finder.fit(adata)
result = finder.select()
```

# Genes for cluster 0

```python
print("weights:", result.weights[0][:5])
sc.pl.umap(adata, color=result.genes[0][:5])
```

```python
print("weights:", result.weights[2][:5])
sc.pl.umap(adata, color=result.genes[2][:5])
```
